## Introduction

Online courses often do not involve face-to-face instruction so the uninitiated can easily believe that cheating is easy. 
After all, how could a professor that is miles or even states away prevent students from just googling the answers to their tests? 

Firstly we want to analyse the problem of cheating during the student's enrollment on an online exam. When a student wants to enroll on an online exam, he has to insert personal information, like name ,address, birth date and so on. And he also has to upload a picture or a pdf of an official document, like a passport , an id document .

The question is, how can we  verify that this data is correct?

For instance the identity may not exist, or the name or the birth date may be falsificated or the picture of the official document or the photo in the official document may be  modificated. 
For example, a student may change the photo on the document to make another person take the exam in his place: we can check if the photo has been modified by looking at the edge and verify if the stamps are official.
We can register the keyboard pattern of the student from the words he types during enrollment and we create a biometric profile: during the exam we check if the student who is typing matches the recorded profile.

We assume that the student has access to a personal computer , for now to simplify our work we assume that every student has a laptop with an integrated webcam, a keyboard and a microphone.

Before the exam we ask the student to follow some rules:

-  remove every USB devices from his personal computer, like mouse, external keyboard, hard disk, USB stick and so on and so forth;
-  do not use phone or other devices during the exam;
-  do not look at books and notes;
- during the exam do not switch tabs or close this window or open a document or a browser;
-  remove headphones or other ‘foreign’ objects like hat , scarf,... ;
-  do not talk with other people during the exam;
-  keep the webcam in front of his face;

We also will put a time limit to finish the exam, so the student doesn't have time to cheat.
We will now analyse all the possible strategies in detail.

## 2. Document verification

Giulia Fontanella

**2.1 INTRODUCTION**

Identity document verification is an important part of sensitive processes like enrollment and registration. Reliable verification is vital to combating identity fraud and related criminal activities: identity theft profoundly affects the public and private sectors worldwide and clearly there is a growing need for reliable identity document verification, to enable public and private organizations to combat fraud. For this reason, many software that check the authenticity of a document were developed and we can apply them in our project to identify students during online exams.

Many types of document fraud exist:
-  Counterfeiting: it is the complete reproduction of an identity document. This is simple to detect as counterfeits are often of very poor quality and rarely include all security elements present in an authentic document.
-  Forgery: this involves changes made to an existing document. Changes might be simple ones, like changing the birthdate or expiration date, or more complex. 
-  Stolen blank document. When the job of personalization is decentralized to local authorities, blank ID documents are sent to these points. Forgers take advantage of this transit to steal some of the blanks, and then personalize them to produce what is commonly called a “stolen blank.” 
-  Identity theft: identity theft is impossible to detect by simply checking an ID document, as the act involves a real ID presented by someone who is not the legal bearer. In this case a cross-check is essential.

We are interested in some kind of forgery that can be employed by students on online tests: students can falsify a document to make someone else take the exam in their place, for example, so they may change the photo. They could also change their birthdate or some other data for some reason. 


<img src="g1.jpg">

Identity document verification lets one confirm the legitimacy of an ID through automatic or manual checks on the document’s authenticity. In particular, this process verifies various security elements within the documents. The process can be performed:
-  On the ID itself: it allows more precise checks and the use of specialized tools (UV light, etc.)
-  On one or more images of the document: this is the principal approach when verification must be done remotely (online verification) or when verification is delegated to a specialized, centralized team


For our project we have to use the second approach and check the document remotely. Major differences exist between automatic ID document authentication services offered online. Now we will analyze two of these software that we can implement for our anti-cheating project.

## 3. Keyboard pattern recognition

Giulia Fontanella

**3.1 INTRODUCTION**

Keystroke dynamics refers to the automated method of identifying or confirming the identity of an individual based on the manner and the rhythm of typing on a keyboard. It is a behavioral biometric, this means that the biometric factor is something the person does in an unique way, such data cannot be imitated. 

Nowadays it is considered as an effective methods of online security, used as a secondary means of authentication. 

For our project it will work following these steps:
-  Enrollment: during student enrollment, we record the typing pattern for one short sentence.
-  Authentication: during course or examination, new typing samples are recorded and matched against the enrolled pattern.
-  Access: the software instantly tells if students are who they claim to be.


**3.2 HOW IT WORKS**

Keyboard dynamics analyzes a user's typing patterns. The speed of typing is not so reliable, it can vary a lot because of tiredness, position of the user, type of keyboard, etc.  Basically, keyboard dynamics analyzes the dwell time and flight time and creates a unique biometric profile that is used to identify a user. 

-  Dwell time is the time duration that a particular key is pressed. This gives rise to a lot of variations: for example, you may press the ASDFJKL keys very fast since it is on the natural keyboard finger positions, however, the ZXCV keys are obscured keys, so here the difference becomes most distinct. 
-  Flight time is the time duration in between releasing a key and pressing the next key 

Dwell time and flight time together make the main pattern of the typing rhythm, this is the main part of the signature typing pattern. When typing a series of characters, flight time and dwell time are specific to that subject, and can be calculated in such a way that it is independent of overall typing speed. 

The rhythm with which some sequences of characters are typed can be very person dependent. For example someone used to typing in english will be quicker at typing certain character sequences such as ‘the’ than a person with french roots.

Another factor to keep in mind is the mistake-making pattern. Different people have different ways to approach and correct their typos. Some people don’t care, some use backspace and delete the whole thing and retype, some use a mouse and click on the mistake site, some people use arrows to move to the mistake position.





**3.3 APPLICATIONS AND DEVELOPED SOFTWARE**

Keystroke dynamics can be used for authentication. Actually it is used mostly together with user ID and password credentials as a form of multifactor authentication. In fact you can use it along with other factors to make guesses about whether user in one session is the same as in another.

Another use is as a very specific form of surveillance. There exist software solutions which, often without end-users being aware of it, track keystroke dynamics for each user account. This tracking, historization of keystroke dynamics is then used to analyse whether accounts are being shared or in general are used by people different from the genuine account owner. Reasons for such an implementation could be verification of users following security procedures (password sharing) or to verify that no software licenses are being shared.

In 2013, Coursera began using typing biometrics as part of the authentication process, which made them one of the first companies to do so. The major universities that have partnered with the free online education resource wanted their materials protected by some form of enhanced login authentication. 

One of the largest online proctoring companies, ProctorU, began utilizing this technology in 2015. This decision was made to help reduce issues such as financial aid fraud. At a university level, typing biometrics has become a filter that is used to verify that students are the ones logging in and taking tests. This doesn’t necessarily replace proctoring altogether, but it aids the verification process. 
Numerous other traditional proctoring companies that want to automate their services are now looking at typing biometrics as an easy solution.

We found some companies which develop software products applying keystroke dynamics:

-  *TypingDNA* built and AI engine able to match any two typing patterns with high accuracy. Its authentication API is suitable for securing logins, enforcing reset passwords, detecting intruders and online biometric authentication for user behaviour analytics, multifactor authentication, user identification, eLearning and fraud prevention. They also provide a continuous authentication app, for Windows and Mac, also based on keystroke dynamics. The main “typing pattern” that the software can record is an array of numbers: keystroke dynamics statistics about 44 keys that are used most of the time. Every time the software collects a typing pattern, we get a 320 feature vector regardless of the number of pressed keys. 

-  *ID Control* is a dutch company developing authentication solutions, some of which use keystroke dynamics. 

-  *BehavioSec* is a swedish company specialized in continuous authentication systems: it is a software which monitors activity on a computer to make sure that it is the genuine account owner who is using the computer. BehavioSec uses not only keystroke dynamics but also mouse dynamics and the general way in which the user interacts with the computer: it combines keystroke dynamics with other interactions, such as mouse movements (acceleration time, click frequency, etc).

To have the best accuracy one should be aware of how text statistics are used in typing biometrics. 

*What makes a typing pattern more accurate?*

-  Many different and rare characters: when you have a complex text (15-25 different chars), with a good percentage of usage per character, your user profile accuracy increases. Rare keys (q,x,z) seem to hide higher differences in typing.
-  Longer texts help ensure more accurate statistics about the pressed keys, helping to record a better user profile. For enrollment we recommend recording 160-200 chars of text, but for matching afterwards you can have as few as 90 chars (120-140 is recommended).
- Focus on lower case letters only, as more natural typing happens on lower case letters, compared to numbers and symbols.

*What can decrease accuracy?*

-  If the texts recorded for enrollment and verification contain a lot of similar or identical words, users may type in similar ways and be harder to distinguish. People have a natural tendency to type more similarly when letters come in the same order.
-  With very short texts you typically increase the false rejection rate/false negatives, however since there are fewer chars being recorded it may also lead to a slightly higher false acceptance rate (false positives). It is recommended to use a lower quality threshold for verification when you use shorter texts (but keep in mind that this will increase false positives a bit). Note that we only use typing biometrics authentication as a second factor. Since we don’t really need an optimized false positive rate, we can still use short texts and a lower quality threshold.
-  Avoid using texts with an average of less than 15 different chars, even if the whole text is longer than usual. Since we capture fewer different chars we’ll get less accuracy in matching that will result in more false positives and more false negatives.

**3.5 ADVANTAGES AND POSSIBLE ISSUES**

In general, behavioral biometrics such as keystroke dynamics are less reliable than physiological biometrics. We use various criteria to evaluate the suitability of keystroke dynamics.

-  This biometric solution can be used by all individuals that are able to use a keyboard. There is no special hardware needed as with other biometrics: it is suitable for our project since it requires only a standard computer keyboard. It is also possible to capture the keyboard dynamics in the background, without causing any additional overhead for the user.

-	Unlike physiological biometric factors, there can't be an absolute match with behavioral biometrics. Any two typing patterns are different, even those from the same individual: while no one else can type like you on a keyboard, you will never type identically twice. So in order to perform typing biometrics authentication you need to rely heavily on machine learning. Matching a pattern against others preserving high accuracy, high speed, and low cost is not an easy job.

-  A major problem with keystroke dynamics is that a subject’s typing rhythm varies considerably in between days and even within the same day. There are numerous reasons for this variation, for example it could be caused by tiredness, switching computers or keyboards, mood, influence of alcohol and medications, etc. The effectiveness of this software has high variations because it depends on these external factors.

-	Depending of the country you are in, using key logging software might be a violation of local laws. Even if the actual typed text is not analyzed or retained, applicable legislation is in your disadvantage when you intend to use keystroke dynamics. 

-	It is almost impossible to mimic another person’s typing rhythm. But electronically capturing using keylogging software is possible, thus implementing this biometric solution requires that data security is guaranteed from the input to the matching algorithm, otherwise someone else could steal your identity by copying your pattern.

Model training aims to learn an embedding $f(x)$ of image $x$ such that the squared L2 distance between all faces of the same identity is small and the distance between a pair of faces from different identities is large. This can be achieved with a *triplet loss* $L$ that is minimized when the distance between an anchor image $x^a_i$ and a positive image $x^p_i$ (same identity) in embedding space is smaller than the distance between that anchor image and a negative image $x^n_i$ (different identity) by at least a margin $\alpha$.

$$L = \sum^{m}_{i=1} \large[ \small {\mid \mid f(x_{i}^{a}) - f(x_{i}^{p})) \mid \mid_2^2} - {\mid \mid f(x_{i}^{a}) - f(x_{i}^{n})) \mid \mid_2^2} + \alpha \large ] \small_+$$

$[z]_+$ means $max(z,0)$ and $m$ is the number of triplets in the training set. The triplet loss in Keras is best implemented with a custom layer as the loss function doesn't follow the usual `loss(input, target)` pattern. This layer calls `self.add_loss` to install the triplet loss:

```python
from keras import backend as K
from keras.models import Model
from keras.layers import Input, Layer
```

```python
from keras import backend as K
from keras.models import Model
from keras.layers import Input, Layer

# Input for anchor, positive and negative images
in_a = Input(shape=(96, 96, 3))
in_p = Input(shape=(96, 96, 3))
in_n = Input(shape=(96, 96, 3))

# Output for anchor, positive and negative embedding vectors
# The nn4_small model instance is shared (Siamese network)
emb_a = nn4_small2(in_a)
emb_p = nn4_small2(in_p)
emb_n = nn4_small2(in_n)

class TripletLossLayer(Layer):
    def __init__(self, alpha, **kwargs):
        self.alpha = alpha
        super(TripletLossLayer, self).__init__(**kwargs)
    
    def triplet_loss(self, inputs):
        a, p, n = inputs
        p_dist = K.sum(K.square(a-p), axis=-1)
        n_dist = K.sum(K.square(a-n), axis=-1)
        return K.sum(K.maximum(p_dist - n_dist + self.alpha, 0), axis=0)
    
    def call(self, inputs):
        loss = self.triplet_loss(inputs)
        self.add_loss(loss)
        return loss

# Layer that computes the triplet loss from anchor, positive and negative embedding vectors
triplet_loss_layer = TripletLossLayer(alpha=0.2, name='triplet_loss_layer')([emb_a, emb_p, emb_n])

# Model that can be trained with anchor, positive negative images
nn4_small2_train = Model([in_a, in_p, in_n], triplet_loss_layer)
```

During training, it is important to select triplets whose positive pairs $(x^a_i, x^p_i)$ and negative pairs $(x^a_i, x^n_i)$ are hard to discriminate i.e. their distance difference in embedding space should be less than margin $\alpha$, otherwise, the network is unable to learn a useful embedding. Therefore, each training iteration should select a new batch of triplets based on the embeddings learned in the previous iteration. Assuming that a generator returned from a `triplet_generator()` call can generate triplets under these constraints, the network can be trained with:

```python
import numpy as np
import os.path

class IdentityMetadata():
    def __init__(self, base, name, file):
        # dataset base directory
        self.base = base
        # identity name
        self.name = name
        # image file name
        self.file = file

    def __repr__(self):
        return self.image_path()

    def image_path(self):
        return os.path.join(self.base, self.name, self.file) 
    
def load_metadata(path):
    metadata = []
    for i in os.listdir(path):
        for f in os.listdir(os.path.join(path, i)):
            # Check file extension. Allow only jpg/jpeg' files.
            ext = os.path.splitext(f)[1]
            if ext == '.jpg' or ext == '.jpeg':
                metadata.append(IdentityMetadata(path, i, f))
    return np.array(metadata)

metadata = load_metadata('images')
```

### Distance threshold

To find the optimal value for $\tau$, the face verification performance must be evaluated on a range of distance threshold values. At a given threshold, all possible embedding vector pairs are classified as either *same identity* or *different identity* and compared to the ground truth. Since we're dealing with skewed classes (much more negative pairs than positive pairs), we use the [F1 score](https://en.wikipedia.org/wiki/F1_score) as evaluation metric instead of [accuracy](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html).

```python
from sklearn.metrics import f1_score, accuracy_score

distances = [] # squared L2 distance between pairs
identical = [] # 1 if same identity, 0 otherwise

num = len(metadata)

for i in range(num - 1):
    for j in range(1, num):
        distances.append(distance(embedded[i], embedded[j]))
        identical.append(1 if metadata[i].name == metadata[j].name else 0)
        
distances = np.array(distances)
identical = np.array(identical)

thresholds = np.arange(0.3, 1.0, 0.01)

f1_scores = [f1_score(identical, distances < t) for t in thresholds]
acc_scores = [accuracy_score(identical, distances < t) for t in thresholds]

opt_idx = np.argmax(f1_scores)
# Threshold at maximal F1 score
opt_tau = thresholds[opt_idx]
# Accuracy at maximal F1 score
opt_acc = accuracy_score(identical, distances < opt_tau)

# Plot F1 score and accuracy as function of distance threshold
plt.plot(thresholds, f1_scores, label='F1 score');
plt.plot(thresholds, acc_scores, label='Accuracy');
plt.axvline(x=opt_tau, linestyle='--', lw=1, c='lightgrey', label='Threshold')
plt.title(f'Accuracy at threshold {opt_tau:.2f} = {opt_acc:.3f}');
plt.xlabel('Distance threshold')
plt.legend();
```

#### 6.3.1 Dlib Pipeline
We are now going to introduce the pipeline needed to create an objet detection framework. The first problem that we have to deal with is to scan the image in order to localise the object, regardless of the position and the dimension of the object. The scanner can be divided into two different components: an image pyramid and a sliding window.
##### Image Pyramid
An image pyramid is just a multi-scale representation of an image. This allows us to find objects at different scales. At the bottom of the pyramid we have the original image at its original size, while at each of the other layers the image is resized (subsampled) and smoothed through a Gaussian blurring. Let’s see how to implement an image pyramid using the OpenCV library.

```python
import imutils

def pyramid(image, scale=1.5, minSize=(30,30)):
    #yeld the original image
    yield image
    
    #looping over
    while True:
        #compute the new dimensions and resize
        w=int(image.shape[1]/scale)
        image=imutils.resize(image, width=w)
        if image.shape[0]<minSize[1] or image.shape[1]<minSize[0]:
            break
    yield image
```

##### Sliding Window
A sliding window is a rectangular region of fixed width and height that slides across an image. For each of these windows we would take the window region, extract features format and apply an image classifier to determine if the window contains an object of interest. Combined with the previously seen image pyramid technique, we can create a classifier which is able to recognise object at various scales and positions.
Notice that varying the value of the two parameters we are able to modify the shape of our bouding boxes, in order to adequate them to the object we are looking for, and adjust the dimension of the slide between the various boxes, in order to reduce the computational complexity (despite reducing also the probability of recognition): in practice it is common to use a sliding window between 4 and 8 pixels.
Here we can see an easy implementation

```python
def sliding_window(image, stepSize, windowSize):
    #slide across the image
    for y in range(0, image.shape[0], stepsize):
        for x in range(0, image.shape[1], stepsize):
            yield (x, y, image[y:y+windowSize[1], x:x+windowSize[0]])
```

#### 6.4 Building the framework
Despite OpenCV comes with methods to train our own custom Haar classifier, they have some problem which we want to avoid:
-  Haar cascades are slow to train
-  Haar cascades have an high false-positive rate
-  Parameters can be hard to tune

For these reasonds the Haar cascade models are not frequently used and they have been substituted by an approach which takes advntage of Histogram of Oriented Gradients (HOG) and Linear Support Vector Machine (LSVM).
We'll now introduce the steps required to build our custom object detector.
##### Sampling
We sample P positive samples from our training data and use the extracted bounding boxes to compute their HOG descriptor.
We sample N negative samples and compute their HOG descriptor.
Note that in order to achieve a good permormance we should take N>>P.
##### Training
Train a Linear SVM on our samples.
##### Hard Negative Mining
For each image in our negative training set we apply the image pyramid and the sliding window techniques.
For each window we then compute the HOG descriptor and apply the classifier. Each time the classifier classifies the current window as an object (false positive) we record the feature vector associated and use it to improve our classifier, re-training it and reducing the final number of false positive.
This step can be repeted many times.
##### Testing
After scanning all the image, we select the bounding boxes wich a sufficiently large probability and apply non-maximum suppression to remove redundant and overlapping boxes.
Thanks to an extremely accurate algorithm used to find the hyperplane separating the two image classes, Dlib obtains a higher accuracy compared to other state-of-the-art detector.

#### 6.5 Histogram of Oriented Gradient
Histogram of Oriented Gradients, or HOG for short, are descriptors mainly used in computer vision and machine learning for object detection. HOG features were first introduced by Dalal and Triggs in their CVPR 2005 paper, Histogram of Oriented Gradients for Human Detection. In their work, Dalal and Triggs proposed HOG and a 5-stage descriptor to classify humans in still images.
These 5 stages are:
-  Normalizing the image prior to description
-  Computing gradients in both the x and y directions
-  Obtaining weighted votes in spatial and orientation cells
-  Contrast normalizing overlapping spatial cells
-  Collecting all HOG to form the final feature vector

The most important parameters for the HOG descriptor are:
-  orientations
-  pixels_per_cell
-  cells_per_block

These three parameters effectively control the dimensionality of the resulting feature vector. In most real-world applications, HOG is used together with a Linear SVM to perform object detection. The reason HOG is frequently utilized is because local object appearance and shape can be characterized using the distribution of local intensity gradients.
HOG descriptors are used to describe the structural shape and appereance of an object in an image, returning a real-valued feature vector.

#### How do the descriptor work?
The fundamental on which HOG descriptor algorithm are based is the fact that appearance of an object can be modeled by the distribution of intensity gradients inside rectangular regions of an image.
Implementing this descriptor requires dividing the image into cells and for each one computing a histogram of oriented gradients for the pixels within each cell. We can then accumulate these histograms across multiple cells to form our feature vector.
Below there is the sample code to extract the HOG description of the following image.
<img src="chess.jpg">

```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt


class Hog_descriptor():
    def __init__(self, img, cell_size=16, bin_size=8):
        self.img = img
        self.img = np.sqrt(img / float(np.max(img)))
        self.img = self.img * 255
        self.cell_size = cell_size
        self.bin_size = bin_size
        self.angle_unit = 45
        assert type(self.bin_size) == int, "bin_size should be integer,"
        assert type(self.cell_size) == int, "cell_size should be integer,"
        assert type(self.angle_unit) == int, "bin_size should be divisible by 360"

    def extract(self):
        height, width = self.img.shape
        gradient_magnitude, gradient_angle = self.global_gradient()
        gradient_magnitude = abs(gradient_magnitude)
        cell_gradient_vector = np.zeros((height // self.cell_size, width // self.cell_size, self.bin_size))
        for i in range(cell_gradient_vector.shape[0]):
            for j in range(cell_gradient_vector.shape[1]):
                cell_magnitude = gradient_magnitude[i * self.cell_size:(i + 1) * self.cell_size,
                                 j * self.cell_size:(j + 1) * self.cell_size]
                cell_angle = gradient_angle[i * self.cell_size:(i + 1) * self.cell_size,
                             j * self.cell_size:(j + 1) * self.cell_size]
                cell_gradient_vector[i][j] = self.cell_gradient(cell_magnitude, cell_angle)

        hog_image = self.render_gradient(np.zeros([height, width]), cell_gradient_vector)
        hog_vector = []
        for i in range(cell_gradient_vector.shape[0] - 1):
            for j in range(cell_gradient_vector.shape[1] - 1):
                block_vector = []
                block_vector.extend(cell_gradient_vector[i][j])
                block_vector.extend(cell_gradient_vector[i][j + 1])
                block_vector.extend(cell_gradient_vector[i + 1][j])
                block_vector.extend(cell_gradient_vector[i + 1][j + 1])
                mag = lambda vector: math.sqrt(sum(i ** 2 for i in vector))
                magnitude = mag(block_vector)
                if magnitude != 0:
                    normalize = lambda block_vector, magnitude: [element / magnitude for element in block_vector]
                    block_vector = normalize(block_vector, magnitude)
                hog_vector.append(block_vector)
        return hog_vector, hog_image

    def global_gradient(self):
        gradient_values_x = cv2.Sobel(self.img, cv2.CV_64F, 1, 0, ksize=5)
        gradient_values_y = cv2.Sobel(self.img, cv2.CV_64F, 0, 1, ksize=5)
        gradient_magnitude = cv2.addWeighted(gradient_values_x, 0.5, gradient_values_y, 0.5, 0)
        gradient_angle = cv2.phase(gradient_values_x, gradient_values_y, angleInDegrees=True)
        return gradient_magnitude, gradient_angle

    def cell_gradient(self, cell_magnitude, cell_angle):
        orientation_centers = [0] * self.bin_size
        for i in range(cell_magnitude.shape[0]):
            for j in range(cell_magnitude.shape[1]):
                gradient_strength = cell_magnitude[i][j]
                gradient_angle = cell_angle[i][j]
                min_angle, max_angle, mod = self.get_closest_bins(gradient_angle)
                orientation_centers[min_angle] += (gradient_strength * (1 - (mod / self.angle_unit)))
                orientation_centers[max_angle] += (gradient_strength * (mod / self.angle_unit))
        return orientation_centers

    def get_closest_bins(self, gradient_angle):
        idx = int(gradient_angle / self.angle_unit)
        mod = gradient_angle % self.angle_unit
        if idx == self.bin_size:
            return idx - 1, (idx) % self.bin_size, mod
        return idx, (idx + 1) % self.bin_size, mod

    def render_gradient(self, image, cell_gradient):
        cell_width = self.cell_size / 2
        max_mag = np.array(cell_gradient).max()
        for x in range(cell_gradient.shape[0]):
            for y in range(cell_gradient.shape[1]):
                cell_grad = cell_gradient[x][y]
                cell_grad /= max_mag
                angle = 0
                angle_gap = self.angle_unit
                for magnitude in cell_grad:
                    angle_radian = math.radians(angle)
                    x1 = int(x * self.cell_size + magnitude * cell_width * math.cos(angle_radian))
                    y1 = int(y * self.cell_size + magnitude * cell_width * math.sin(angle_radian))
                    x2 = int(x * self.cell_size - magnitude * cell_width * math.cos(angle_radian))
                    y2 = int(y * self.cell_size - magnitude * cell_width * math.sin(angle_radian))
                    cv2.line(image, (y1, x1), (y2, x2), int(255 * math.sqrt(magnitude)))
                    angle += angle_gap
        return image

img = cv2.imread('notebook_images/chess.jpg', cv2.IMREAD_GRAYSCALE)
hog = Hog_descriptor(img, cell_size=8, bin_size=8)
vector, image = hog.extract()
plt.imshow(image, cmap=plt.cm.gray)
plt.show()
```

#### 6.6 Support Vector Machine
Support Vector Machine (SVM) is a popular supervised learning models associated with a learning algorithms that analyze data used for classification and regression analysis. In this algorithm, each data item is plot as a point in a n-dimensional space (n representing the number of features) in which the value of each feature is the value of a particular coordinate. The classification is then performed by finding the hyperplane which best differentiate the two classes.
The reason SVM is so popular is because it has solid theoretical foundations and has good out-of-the-box performances. There are still hyperparameters to tune, but throwing an SVM at a problem is a good quick way to get a baseline accuracy with minimal parameter tuning; however, to achieve a good accuracy tuning is needed.

#### 6.7.5 Webcam stream detection
Since we are interested in detecting the presence of the headphones during an online examination, we want to apply the detector to the webcam video stream. We used openCV to capture the video from the webcam and then, to reduce to computational complexity, we sample an image every 0.1 seconds (10fps) and apply the detector to that picture. If headphones are detected the frame is saved in the "detected" folder, in this way one coul double check manually that the person is really cheating and we are not occurring in a false positive.
The implemetation below is based only on the headpones detector, without considering the airpods. This is due to the fact that the low definition of the webcam does not permit it to properly recognize them.

```python
from imutils import paths
import argparse
import dlib
import cv2
import numpy as np
import time

#construct the argument parser and parse the argument
detector="detector/detector.svm" #Path to trained object detector

#Set the duration of the exam in minutes
duration=1
tot=duration*60
start_time=time.time()

#Start video capture
cap = cv2.VideoCapture(0)
fps=10

#load the detector
detector=dlib.simple_object_detector(detector)
#loop over the testing images
while(time.time()-start_time<tot):
    #load the image and make predictions
    time.sleep(1/fps)
    ret, frame = cap.read()
    boxes=detector(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    #loop over the bounding boxes and draw them
    for b in boxes:
        (x,y,w,h)=(b.left(),b.top(),b.right(),b.bottom())
        cv2.rectangle(frame,(x,y),(w,h),(0,255,0),2)
        #if the object is detected save the frame
        if (b.left()):
            cv2.imwrite('detected/image{}.jpg'.format(time.time()),frame)

    #show the image
    cv2.imshow("Image",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
def _find_device(self, idVendor, idProduct):
        """Find a USB device by product and vendor id."""
        for bus in usb.busses():
            for device in bus.devices:
                if (device.idVendor == idVendor and
                    device.idProduct == idProduct):
                    return device
        return None
```

### Introduction
In this project, it will be evaluated, if gaze detection is suitable to prevent cheating in online test environments.
Background here is the digitalisation of educational institutions allowing students to take tests, do homework and get certifications without being present at the institution. This is achieved by giving students a program that is accessible via internet in which the tasks for the examination are implemented and can be answered by the testee.

The problem emerges when graded tests are taken and significant certificates are at stake like admission tests for universities or exams in those universities. When there is noone present to look out for cheaters, an online solution has to be found to ensure the participants abide by the rules. 

This part of the 'Prevent online cheating' project is about the possible usage of gaze detection to prevent people from looking away from the screen to cheat.

### Goal

To prevent cheating with gaze detection, we first have to find out which ways there are to cheat via looking somewhere else.

   1. Pinning answers to a wall behind/ next to the screen     
    
    
   2. Putting answers near the keyboard
    
    
   3. Placing answers onto the screen via paper
    
    
Rather than having a cheat sheet there, the sheets could be replaced by a cellphone or any other item that gives information or even a person that shows answers to the testee. 

The initial idea is to use gaze detection to find out when, how long or how often the testees gaze leaves the computer screen to look at the cheat sheets and penalize him on the base of this observation.

### Gaze detection basics

Gaze detection is a technique used by computers to analyze a video input for a persons direction of looking. First it analyzes the video to recognize the face of the user and to localize its pupils. To compute the direction in which the user looks, calibration has to take place first. Here, different dots are displayed on the screen,b that the subject has to klick on while looking at them to help the program map out the screen according to the distance and direction from which the subject is looking at it. This is done by measuring the distance between the pupils and the inner corners of the eye. Depending on the distance, the direction of the gaze can be determined. 

### Problems

There are several problems that have to be taken into account when using eye tracking as a way to prevent cheating in online tests. Some of the problems can be solved by having people taking the test meeting certain prerequisites. This means that they need to at least have a camera for example. 

##### Accuracy

The quality and accuracy of the gaze detection tools greatly depend on the quality of the camera used by the testee. In our project prerequisites, we decided that every person taking part at least needs a laptop with a camera and certain other equipment (that is not further relevant for this part of the project). Although this means that everyone is able to use the gaze detection tool, it is not a guarantee for a qualitative metric to decide if someone is cheating. 
First and foremost, the gaze detection is very inaccurate, when only a standard laptop camera is used. 
VIDEO/GIF OF GAZE DETECTION GAME
For our testees, that would mean that anytime their gaze is moving close to the border set by the screen, there is a good chance, the gaze detection would give the feedback, that they are looking at something outside the screen, which is a considerable hinderance for them, since the testees eye-movement is already restricted to the screen.

But the accuracy of the gaze-measurement is not solely reliant on the video quality but also on the surroundings of the person. If the light is too bright or too dim, the gaze detection is worse aswell, which would mean, that everyone would have to have a well lit face in order to have the face and eyes correctly recognized.

Last but not least, there is another important factor that makes gaze detection considerably more inaccurate and fault-prone which are glasses or contact lenses. By nature, having something in the way between the camera and the eyes, the measuring accuracy becomes worse, but with glasses there might be positions where the pupils are obscured by the frame of the glasses. Wearing contact lenses also doesn't solve the problem since they do reflect light, also somewhat less than glasses, which interferes with the detection.
This poses a prerequisite for testees to take off their glasses or at least wear contact lenses, but since that might well be a prerequisite that is not fulfillable by a great enough number of people to make gaze detection not viable for detecting cheaters.

An alternative would be using an eye tracking device like the Tobii eye tracker [1] or an eye tracking glass. Those devices project a small red dot in the eyes of the user and use them as a stable measuring point instead of the eye-corner. This makes the eye tracking a lot more accurate, but expecting everyone to buy suach a device which cost more than 100€ is not a feasable option.

##### Movement restrictions

There are even more restrictions placed on a testee when they have to use the gaze detection system:

They need to sit straight and move as little as possible to stay in focus of the camera for the whole duration of the tests. Since online examinations might take several hours to complete, this might become very uncompfortable after even a short while.

They can, under no circumstances look away from the screen, which means that you cannot just take a short break from the test while looking away from the screen to get a bit of rest for the eyes. You can not look away from the screen when you eat or drink something when taking the test.
And most importantly, you can not look at your keyboard to write or correct mistakes, which means that noone who can not write blindly could take the tests. 
This means that either you have to allow people to look away from the screen once in a while and be able to look at their keyboard or you try to have the camera also monitor the keyboard to make sure that there is nothing but a keyboard there. This however does not work since the camera will not be able to capture both the face and the keyboard as a whole especially on a laptop. Allowing people to take a break and maybe implementing a "pause" system would make the whole anti-cheat system obsolete. This means that all the restrictions named would be necessary for gaze detection to be properly used.

##### Calibration

There is no way of knowing how the computer-setup of a testee looks like. Although the project prerequisites say that there needs to be a laptop. The laptop could have displays of various sizes.
The calibration of the gaze detection software works by displaying dots on the screen, especially at the edges, and the user has to look at those dots and confirm that spot (for example by pressing enter) while looking at it, so the software can map out the display and compute whether the user is looking at the screen.
 [4]
The problem occurs when people know how to cheat at calibration, which is pretty easy. When the borders of the screen are mapped out in the calibration phase, you can just look outside the screen (a bit above and a bit to the sides) to simulate a bigger screen. The system will then assume that you are still looking at the screen, when in reality you gaze above it to see the cheat sheet. Since there is no way to map out the screen in relation to the testee, calibration is a necessity. Ways to prevent this pretty easy and obvious way of cheating the calibration are not obvious.

### Implementation

The goal was to implement a basic gaze-detection tool and modify that so that it would give feedback when the gaze of the user leaves the screen.


#### PyGaze [2]

The first approach in implementing a simple gaze detection tool was started with Pygaze, a Python based framework to implement different kinds of gaze detection. Using the complete package that is available for windows, an EDA named Spyder has been installed together with the required packages to start some basic tutorial projects in Pygaze. This took a considerable while since although it is prepared to run after installation, the display classes could not automatically detect the dimensions of the screen automatically. Manual correction of the problem did not solve the problem since dimensions that should fit the screen or were even smaller did not produce a fully visible screen. Despite of this, a test can be run that allows calibration by simulating the gaze of a person with the mouse-cursor.
That test was all that could be run since the other tests resulted in errors that could not be easily fixed with nearly no knowledge of the framework nor python.

Furthermore, PyGaze itself does not provide any tools to connect to the camera directly. Another framework would have had to be used to feed PyGaze with video-information. One such framework would is provided by Tobii aswell, but only works with their tools.
Since PyGaze itself could not be used properly due to having no expirience with that framework nor with the video framework, an implementation this way would have taken far too much time to be acceptable in this project. 

#### WebGazer [3]

Webgazer is a browser based eye tracking tool that can be incorporated into a website via an easy link. Although incorporationg a link on a website would have been relatively easy, modifying that tool to recognize the gaze leaving the screen would have been difficult since there is no experience with java script aswell. 

#### Installation summary

Due to complication and inexpirience with the PyGaze framework, an implementation of a gaze detection tool that gives feedback when the user's gaze leaves the screen was unsuccessful and was excluded from the project due to time constraints caused by delays in the formulation of the project goals.


### Summary

Researching on the topic of using gaze detection as a means to prevent cheating at online tests,It has become clear that the restrictions that have to be put on testees to use the tool are too much of a burden to still allow participation in an exam.
Having to not look up at any time, not being able to look at the keyboard and having to take off glasses or contact lenses restricts too many people from participating. The alternative would be to require a specific eye tracking tool to be owned by each testee is also not feasible since they cost too much. Furthermore, it is too easy to trick the gaze detector by influencing the calibration.
If all those restriction would not be required when taking the exam, the gaze detector would simply not be accurate enough to ensure correct detections of cheaters.

In conclusion, gaze detection in its current form is not an option to combat cheaters. There still is potential for when the detection is less prone to disruptions and is more accurate with standard equipment.

## 9. Conclusions

This paper examines the deployment of a system to block possible cheating during an online exam, one of the core problems of the online courses.
We start to analyse the problem dividing it in 3 sub-categories:
cheating during the enrollment, cheating alone and cheating with another person.
And then we have identified some anti-cheating strategies to be used.
Each chapter analyzes one of them, with the problems encountered, with the constraints we had to assume, and with some implementation.

To prevent cheating during enrollment, we found a program that can verify the authenticity
of a document and can detect if the photo in the document has been modified.
We also found a program that can record the typing pattern of the student during enrollment and match it against the typing pattern during the exam, to verify that the person who is typing is still the same.

For the problem of finding the usb devices we were able to find both a program that lists all the devices attached to the student's computer and we were also able to implement a python code that returns a list with the active usb. So if this list is not empty during the exam, we will suspend the exam.

To prevent the user from interacting with other people, we decided to implement an object detector, capable to spot the presence of headphones through which the user could keep in contact and obtain suggestions. The main issues limiting our results are the tiny dimension of our dataset, the low resolution of the webcam and the absence of a gpu. Despite we managed to successfully identify headphones with a sufficient precision, there are still large margin of improvement.
